```
import pandas as pd
import numpy as np
import scipy
from scipy.sparse import load_npz
from scipy.sparse import csr_matrix
from scipy.sparse import save_npz
from pandas.api.types import CategoricalDtype

from google.colab import drive
drive.mount('/content/drive')
DATADIR = "/content/drive/My Drive/sbermarket_dataset/kirya"
```

```
_co_oc_items_normed = pd.read_csv(f"{DATADIR}/co_oc_items.csv")
_co_oc_items_normed
```

```
df_diag = pd.read_csv(f"{DATADIR}/df_diag.csv")
df_diag.drop(columns=['Unnamed: 0'])
```

```
_co_oc_items_normed = pd.concat([_co_oc_items_normed, df_diag], axis=1)
```

```
_co_oc_items_normed.drop(columns=['Unnamed: 0'])
```

```
_co_oc_items_normed['max_normed'] = _co_oc_items_normed['max'] / _co_oc_items_normed['0'] 
```

```
_co_oc_items_normed['mean_normed'] = _co_oc_items_normed['mean'] / _co_oc_items_normed['0'] 
_co_oc_items_normed['median_normed'] = _co_oc_items_normed['median'] / _co_oc_items_normed['0'] 

_co_oc_items_normed.drop(columns=['Unnamed: 0', '0'])
```

```
_co_oc_items_normed = _co_oc_items_normed.drop(columns=['Unnamed: 0', '0'])
_co_oc_items_normed
```

```
_co_oc_items_normed.to_csv(f"{DATADIR}/co_oc_items_final.csv")
```
